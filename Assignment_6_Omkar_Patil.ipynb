## **1. Print every element inside this nested list:**

## **2. Find the sum of all numbers inside a nested list.**
Example: `[[10,20], [30], [5, 7, 8]]`

````python
# Write your answer here
lst=[[10,20], [30], [5, 7, 8]]
print(lst)
total=0
for i in lst:
    for j in i:
        total+=j

print("sum of all numbers:",total)
````

## **3. Count how many even numbers are present in a nested list.**

````python
# Write your answer here
lst=[[1,2,3], [4,5], [6,7,8,9]]
print(lst)
count=0
for i in lst:
    for j in i:
        if j%2==0:
            count+=1
print("count of even numbers:",count)
````

## **9. Replace all negative numbers with 0 inside a nested list.**

````python
# Write your answer here
lst=[[20,-1,42,59,6,-540,-58,78,20],[55,50,21,-63,10,66,-88,88,4],[3,12,-61,78,5,4,1,100]]
print("old_list:",lst)
for i in range(len(lst)):
    for j in range(len(lst[i])):
        if lst[i][j]<0:
            lst[i][j]=0
    
print("new_list:",lst)
````

````python
#comprehension
ls=[[0 if lst[i][j]<0 else lst[i][j] for j in range(len(lst[i]))] for i in range(len(lst)) ]
ls
````

## **12. Increase every student's marks by 5 in a nested dictionary.**

````python
# Write your answer here
student={'A': {'age': 15, 'grade': 10},'B': {'age': 16, 'grade': 11} }
print("old record:",student)

for i,j in student.items():
    j["grade"]+=5
    
print("new record:",student)
````

````python
#comprehension
student={'A': {'age': 15, 'grade': 10},'B': {'age': 16, 'grade': 11} }

rec = {i: {**j, "grade": j["grade"] + 5} for i, j in student.items()}
rec
````

## **20. Convert all inner dictionary keys to uppercase.**

````python
# Write your answer here
employees = {
    'E1': {'name': 'John', 'status': 'active'},
    'E2': {'name': 'Sara', 'status': 'inactive'},
    'E3': {'name': 'Mike', 'status': 'active'},
    'E4': {'name': 'Rita', 'status': 'inactive'}
}

for emp_id, info in employees.items():
    new_info = {}
    for k, v in info.items():
        new_info[k.upper()] = v     
    employees[emp_id] = new_info

print(employees)
````

````python
#comprehension
employees = {
    'E1': {'name': 'John', 'status': 'active'},
    'E2': {'name': 'Sara', 'status': 'inactive'},
    'E3': {'name': 'Mike', 'status': 'active'},
    'E4': {'name': 'Rita', 'status': 'inactive'}
}

emp={id:{key.upper(): value for key,value in info.items()} for id,info in employees.items() }
emp
````
