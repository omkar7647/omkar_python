### 36. Determine the type of triangle using sides: acute, right, obtuse.

```python
a = float(input("Enter side 1: "))
b = float(input("Enter side 2: "))
c = float(input("Enter side 3: "))
print(a,b,c)
a, b, c = sorted([a, b, c])
if a + b <= c:
    print("Not a valid triangle")
elif c*c == a*a + b*b:
    print("Right Triangle")
elif c*c < a*a + b*b:
    print("Acute Triangle")
else:
    print("Obtuse Triangle")
```

### 42. Check if a number is perfect (sum of factor  equals number).

```python
n = int(input("Enter a positive number: "))
print(n)
if n <= 1:
    print("Not a perfect number")
else:
    total = 1  # 1 is always a proper divisor for n>1
    i = 2
    import math
    while i * i <= n:
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
        i += 1
    print("Perfect" if total == n else "Not perfect")

```
