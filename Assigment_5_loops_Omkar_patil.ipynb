### 40. Create a dictionary of 5 products and their prices. Increase every product's price by 10% using a for loop

```python
products={"pen":10,"book":100,"bag":300,"table":350,"bench":500,"notebook":50,"compass":60}
print("old record:-",products)

for i,j in products.items():
    products[i]=j+(j*10/100) 
    
print("new records:-",products)
```

### 41. Given a dictionary of 10 items, delete all items whose value is less than 50 using a for loop

```python
item={"pen":10,"book":100,"bag":300,"table":350,"bench":500,"notebook":50,"compass":60,"pencil":5}
print("old:-",item)

to_delete_key=[]
for i,j in item.items():
    if j<50:
        to_delete_key.append(i)
print(to_delete_key)

for key in to_delete_key:
    item.pop(key)
    
print("new:-",item)
```

### 42. Create a dictionary of subjects and marks. Add a new subject and mark. Then use a for loop to print all updated keyâ€“value pairs

```python
student={"PHY":76,"MATHS":85,"CHE":80,"ENG":60}
print("old:",student)

new_rec={"BIO":94,"STATS":88}
student.update(new_rec)
print("new:",student)

print("\nUpdates Records:-")
for i,j in student.items():
    if i in new_rec:
        print(i,j)
```
